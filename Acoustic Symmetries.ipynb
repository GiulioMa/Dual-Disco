# Acoustic Symmetry in Microphone Array Configuration

The diagram showcases a strategic arrangement of a two-microphone system in relation to a base plate, which is delineated into nine distinct platforms labeled P1 through P9, forming a 3x3 grid. This setup is constructed to utilize the inherent axes of symmetry in the polar diagram of the microphones for the purpose of acoustic signal differentiation and source localization.

#### Front Microphone Acoustic Symmetry (Bottom Microphone):
The front microphone, positioned at the bottom of the diagram, capitalizes on a polar response pattern where acoustic waves from platforms that are horizontally opposite from each other are perceived as acoustically identical. Consequently, for this microphone, the following platforms generate identical acoustic signals:
- Platforms P1 and P3.
- Platforms P4 and P6.
- Platforms P7 and P9.

This symmetry is critical for differentiating signals on the horizontal axis when paired with data from the second microphone.

#### Rear Microphone Acoustic Symmetry (Right-side Microphone):
In contrast, the rear microphone, illustrated on the right side of the diagram, is positioned such that the acoustic waves from platforms that are vertically opposite from each other are identical. Consequently, the following platforms share the same acoustic characteristics from the perspective of the rear microphone:
- Platforms P1 and P7.
- Platforms P2 and P8.
- Platforms P3 and P9.

#### Integrating Data for Source Localization:
By intersecting the acoustic data from both microphones, we are able to resolve the ambiguity presented by the symmetrical acoustic signatures. The unique combination of symmetries from both the front and rear microphones allows for the accurate determination of the originating platform for any given acoustic wave.
