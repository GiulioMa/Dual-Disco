```python
import torch

# Load the saved PyTorch database
pytorch_database = torch.load('./Data/torch_data')

# Initialize dictionaries for train and test sets
train_data = {}
test_data = {}

# Modes for training
test_modes = [
    " Conduction2 & Conduction2",
    " Conduction2 & Keyhole2",
    " Keyhole2 & Conduction2",
    " Keyhole2 & Keyhole2"
]

# Iterate over the loaded database
for key, data in pytorch_database.items():
    if data['mode'] in test_modes:
        # Append to train dictionary
        test_data[key] = data
    else:
        # Append to test dictionary
        train_data[key] = data
```

```python
for key, data in test_data.items():
    print(key, data['mode'])
```

```python
for key, data in train_data.items():
    print(key, data['mode'])
```

```python
torch.save(train_data, './Data/train')
torch.save(test_data, './Data/test')
```
